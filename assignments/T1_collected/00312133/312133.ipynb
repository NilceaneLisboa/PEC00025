### 2.2. Dados do Sistema 

Os dados utilizados, para elaboração do sistema podem ser consultados a seguir. 

```python
m  = 1.0       # massa do sistema em Kg 
fn = np.linspace(0.1,10,500) #frequencia natural em  Hz
zt = 0.01      #  amortecimento
u0 = 0.        # deslocamento inicial
v0 = 0.        # velocidade inicial
Td = 1/fn
```

Depois que as propriedades do sistema forem especificadas, o carregamento da etapa unitária pode ser criado como uma ocorrência ''MRPy'' chamando o construtor apropriado:

```python
data = MRPy.from_file('data/compressor', form='invh').zero_mean()
t    = data.t_axis()

plt.figure(1, figsize=(8, 8), clear=True, )


for kX, acc in enumerate(data):
    
    plt.subplot(3, 1 ,kX+1)
    plt.plot(t, acc, color=('m'))


    plt.xlim(0, data.Td); 
    plt.ylim(-15, 15);
    plt.ylabel('acc{0} (m/s^2)'.format(kX+1)) 

    plt.grid(True) 

plt.xlabel('time (s)')
```

Aceleração ao longo do eixo z,


```python
az = MRPy(data[2], data.fs)  
t  = az.t_axis()

plt.figure(2, figsize=(8, 4), clear=True)
plt.plot(t, az[0],color=('m'))

plt.xlim(0, az.Td);  plt.xlabel('time (s)') 
plt.ylim(-50, 50);   plt.ylabel('a_z (m/s^2)') 

plt.grid(True)
```

### 2.3 Integral de Duhamel  <a name="section_3"></a> 


### 2.3.1 Formulção  
     

No estudo das teorias de vibrações, a integral de Duhamel é uma maneira de calcular e modelar como sistemas lineares e estruturas respondem a perturbações externas dependentes do tempo.

$$ u(t) = u_0(t) + \frac{1}{m \omega_{\rm D}} 
                   \int_0^t \exp \left[ -\zeta\omega_{\rm n}(t - \tau) \right] \;
                   \sin \omega_{\rm D} (t - \tau) \; F(\tau) \; d\tau $$



```python
umax = np.zeros(fn.shape)

for k, f in enumerate (fn):

 u_DH = az.sdof_Duhamel(fn, zt, u0, v0)/m
 umax[k] = u_DH.max()
wn = (2*np.pi*fn)
```

### 2.4 Espectro de Pseudoaceleração 

## $$ u_{pico} = u_{max} *\omega_{n}^2 \ $$

```python
apico = wn*wn*umax


plt.figure(3, figsize=(19, 4), clear=True)
plt.plot(Td,apico, color=('m'))

plt.xlim(0,10);      plt.xlabel('time (s)') 
plt.ylim(0, 40);   plt.ylabel('aceleração (m/s²)') 

plt.grid(True) 
```

### 2.5 Periodograma

Periodograma é uma estimativa da densidade espectral de um sinal é a ferramenta mais comum para examinar as características de amplitude vs. frequência.Como alternativa, o módulo MRPy também fornece uma visualização rápida do periodograma do sinal, fazendo:


```python
plt.figure(4, figsize=(8, 4), clear=True)
plt.plot(t, az[0],color=('m'))

plt.xlim(0, az.Td);  
plt.xlabel('time (s)') 
plt.ylim(-15, 15);   
plt.ylabel('a_z (m/s^2)') 

plt.grid(True) 

# Método para o cálculo do periodograma

Saz, fs = az.periodogram()
f       = az.f_axis()

plt.figure(5, figsize=(8, 4), clear=True)
plt.title('Periodograma Dados de Entrada') 
plt.plot(f, Saz[0],color=('m'))

plt.xlim(0, 270);   plt.xlabel('frequencia (Hz)') 
plt.ylabel('S_az (power)') 

plt.grid(True) 
```

Já os dados de saída, são expressos no grafico a seguir:


```python
Su_DH, fs = u_DH.periodogram()

f       = u_DH.f_axis()

plt.figure(6, figsize=(8, 4), clear=True)
plt.title('Periodograma Dados de Saída') 
plt.plot(f, Su_DH[0],color=('R'))

plt.xlim(-1, 4);   
plt.xlabel('frequencia (Hz)') 
plt.ylabel('Su (power)') 
plt.grid(True) 

```
