### 4.1. Reading a data file <a name="section_41"></a> 



```python
data = MRPy.from_file('data/iNVH001', form='invh').zero_mean()

az   = MRPy(data[2,:], fs=data.fs)
f1   = az.plot_time(fig=1, figsize=(8, 6), axis_t=[0, az.Td, -10, +10])
```

### 4.2. Defining system properties <a name="section_42"></a> 



```python
fn = 1.0    # natural frequency in Hz
zt = 0.01   # damping as ratio of critical
u0 = 0.     # initial displacement
v0 = 0.     # initial velocity
```

### 4.3. Integrating by finite differences <a name="section_43"></a> 



```python
ufd = az.sdof_fdiff(fn, zt, u0, v0)
f2  = ufd.plot_time(fig=2, figsize=(8, 6), axis_t=[0, ufd.Td, -0.02, +0.02])
```

### 4.4. Integrating by Duhamel <a name="section_44"></a> 



```python
udh = az.sdof_Duhamel(fn, zt, u0, v0)
f3  = udh.plot_time(fig=2, figsize=(8, 6), axis_t=[0, udh.Td, -0.02, +0.02])
```

### 4.5. Comparing results <a name="section_45"></a> 



```python
H   = MRPy.Heaviside(N=1000, t0=2, Td=10)
f4  = H.plot_time(fig=4, figsize=(8, 6), axis_t=[0,   H.Td,   -0.5,   +1.5])

U1  = H.sdof_fdiff(fn, zt, u0, v0)
f5  = U1.plot_time(fig=5, figsize=(8, 6), axis_t=[0,   U1.Td,   -0.05,   +0.1])

U2  = H.sdof_Duhamel(fn, zt, u0, v0)
f6  = U2.plot_time(fig=6, figsize=(8, 6), axis_t=[0,   U2.Td,   -0.05,   +0.1])

err = MRPy(U2-U1, H.fs)
f7  = err.plot_time(fig=7, figsize=(8, 6), axis_t=[0, err.Td, -0.01, +0.01])
```
