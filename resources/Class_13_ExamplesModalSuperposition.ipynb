## 1. Tuned mass damper <a name="section_1"></a> 


<img src="images/TMDsingle.png" alt="Tuned mass damper" width="400px"/>

The system properties and the tentative damper design are given in the following:


```python
# System properties and damper design

m  =  1000.                      # system mass (kg)
mD =  5.                        # damper mass (kg)

fn =  1.                         # natural frequency (Hz)
k  =  m*(2*np.pi*fn)**2          # system stiffness (N/m)

fD =  1.                         # tuned damper frequency (Hz)
kD =  mD*(2*np.pi*fD)**2         # tuned damper stiffness (N/m)

zt =  0.01                       # same damping ration for both (%)

# Tuned damped system matrices

KD = np.array([[ k+kD, -kD ],    # stiffness matrix
               [  -kD,  kD ]])

MD = np.array([[    m,  0  ],    # lumped mass matrix
               [    0,  mD ]])
```

Damper mass must be something between 2 to 4% of effective system mass.
The higher the mass, the broader the frequency range of damper action. But a too large
damper mass is not economically feasible.

Now, the solution for vibration modes is:

```python
# Uses scipy to solve the standard eigenvalue problem
w2, Phi = sc.eig(KD, MD)

# Ensure ascending order of eigenvalues
iw  = w2.argsort()
w2  = w2[iw]
Phi = Phi[:,iw]

# Eigenvalues to vibration frequencies
wk  = np.sqrt(np.real(w2)) 
fk  = wk/2/np.pi

print('First vibration mode:  {0:5.2f}Hz, [{1:6.3f} {2:6.3f}]'.format(fk[0], *Phi[:,0]))
print('Second vibration mode: {0:5.2f}Hz, [{1:6.3f} {2:6.3f}]'.format(fk[1], *Phi[:,1]))
```

The aimed frequency $f_{\rm n} = 1$Hz has bifurcated into a slightly lower and 
a slightly higher frequencies. In first mode both masses displace to the same
direction, while in second mode they displace to opposite directions.

Now we define a harmonic excitation exactly at the original system frequency,
to be applied on the system mass (damper has no direct excitation):


```python
Td =  64.                        # time discretization parameters
N  =  1024
fs =  N/Td

F0 =  0.001*k                    # quasi-static response is 1mm
t  =  np.linspace(0, Td, N)      # time domain
s  =  np.sin(2*np.pi*1.0*t)      # 1Hz sine
F  = [F0*s, np.zeros(s.shape)]   # loading vector

# Modal force as MRPy object
Fk =  MRPy(np.dot(Phi.T, F), fs=fs)
```

Solving the modal equations with Duhamel, and visualizing results:


```python
uk =  Fk.sdof_Duhamel(fk, zt)             # solve by Duhamel
ut =  MRPy(np.dot(Phi, uk), fs=uk.fs)     # back from modal space

ut.plot_time();

print('Peak displacement at system mass: {0:6.3f}m'.format(np.max(np.abs(ut[0]))))
print('Peak displacement at damper:      {0:6.3f}m'.format(np.max(np.abs(ut[1]))))
```
