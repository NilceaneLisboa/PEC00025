## 1. Introduction <a name="section_1"></a> 

Whenever a function $f(t)$ is defined by means of a differential equation 
and a suitable set of boundary conditions, it may be convenient to 
transform the problem of finding $f(t)$, function of an independent
variable $t$ (time) into a new problem of finding $\overline{f}(s)$,
function of a transformed (independent) variable $s$. 
This transformation may be generically expressed as:

$$\overline{f}(s) = \int_{a}^{b} K(s,t) f(t) dt $$

where the function $K(s,t)$ is called _kernel_ of the transformation.
If the integral limits $a$ and $b$ are finites, the function 
$\overline{f}(s)$ is a _finite transformation_.

The 

onde a fun\c{c}\~{a}o $K(s,t)$ das vari\'{a}veis $s$ e $t$ se denomina 
{\em n\'{u}cleo} (``kernel'') da transforma\c{c}\~{a}o. Quando 
ambos os limites $a$ e $b$ na integral \ref{ea.1.1} s\~{a}o finitos , a 
transformada $\overline{f}(s)$ \'{e} uma {\em transformada finita}. 
A partir de \ref{ea.1.1} pode definir-se uma transforma\c{c}\~{a}o inversa,
que permite obter f(t) quando $\overline{f}(s)$ \'{e} dada, se satisfazem 
as correspondentes condi\c{c}\~{o}es de exist\^{e}ncia. f(t) e 
$\overline{f}(s)$ constituem um {\em par de transformadas}. Existem 
tabelas com pares de transformadas que simplificam consideravelmente a 
tarefa de obten\c{c}\~{a}o da transformada inversa, a qual representa a  
maior dificuldade na resolu\c{c}\~{a}o de equa\c{c}\~{o}es diferenciais 
ou integrais mediante a aplica\c{c}\~{a}o de transformadas integrais.

Neste ap\^{e}ndice ser\~{a}o resumidas as propriedades mais importantes das 
transformadas de Laplace, de grande import\^{a}ncia na solu\c{c}\~{a}o de 
sistemas lineares.


## 2. Formal definition  <a name="section_2"></a> 

Seja $f(t)$ uma fun\c{c}\~{a}o definida para todo $t>0$. A transformada 
de Laplace de $f(t)$ se define como:

\begin{equation}
\label{ea.2.1}
\BRA{f(t)} = \int_{0}^{\infty} e^{-st} f(t) dt
\end{equation}

No que segue ser\'{a} tamb\'{e}m utilizada a nota\c{c}\~{a}o:

\begin{equation}
%\label{ea.2.2}
\pounds \BRA{f(t)} = \overline{f}(s)
\end{equation}

Observa-se que o n\'{u}cleo da transforma\c{c}\~{a}o \'{e} neste caso 
a fun\c{c}\~{a}o:

\begin{equation}
%\label{ea.2.3}
K(s,t) = e^{-st}
\end{equation}
sendo s uma vari\'{a}vel complexa.

A condi\c{c}\~{a}o de converg\^{e}ncia da integral \ref{ea.1.1} imp\~{o}e 
certas restri\c{c}\~{o}es a fun\c{c}\~{a}o $f(t)$. Se existem constantes 
reais tais que, com $Re(s)>a$ se verifique a condi\c{c}\~{a}o:

\begin{equation}
%\label{ea.2.4}
\NOR{ e^{-st} f(t) } < C e^{-(s-a)t}
\end{equation}

e se al\'{e}m disso, em qualquer intervalo finito, $f(t)$ apresenta um 
n\'{u}mero finito de descontinuidade finitas e nenhuma descontinuidade 
infinita, ent\~{a}o $\overline{f}(s)$ existe. A primeira condi\c{c}\~{a}o 
implica que $f(t)$ cres\c{c}a para $t\rightarrow\infty$, menos rapidamente 
que $Ce^{at}$. A segunda que f(t) seja seccionalmente cont\'{i}nua 
{\em (``piecewise continuous'')}.

\end{eqnarray}

Deve assinalar-se que $\delta(t-a)$ n\~{a}o constitui uma fun\c{c}\~{a}o 
no sentido cl\'{a}ssico da an\'{a}lise matem\'{a}tica, em que uma 
fun\c{c}\~{a}o $f(t)$ atribui a cada valor da vari\'{a}vel $t$ dentro de 
um certo dom\'{\i}nio, um valor ou quantidade definida $f(t)$. $\delta(t-a)$ 
deve ser considerada, conseq\"{u}entemente, como uma atribui\c{c}\~{a}o.
Aplicando \ref{ea.2.1} a $(t-a)$ obt\'{e}m-se:

\begin{equation}
%\label{ea.3.6}
\pounds \BRA{\delta(t-a)} = \int_{0}^{\infty} e^{-st} \delta(t-a) dt
\end{equation}

\begin{equation}
%\label{ea.3.7}
\pounds \BRA{ \delta(t-a)} = \lim_{\varepsilon\rightarrow 0} \int_{a}^
{a+ \varepsilon} e^{-st} \frac{1}{\varepsilon} dt
\end{equation}

\begin{equation}
%\label{ea.3.8}
\pounds \BRA{\delta(t-a)} = -\frac{ e^{-as}}{s} \BRA{ \lim_
{\varepsilon \rightarrow 0} \frac{e^{-\varepsilon s}-1}{\varepsilon}}
\end{equation}

\begin{equation}
%\label{ea.3.9}
\pounds \BRA{\delta(t-a)} = e^{-as}
\end{equation}

Se a fun\c{c}\~{a}o impulso est\'{a} aplicada na origem, isto \'{e}, 
quando $a=0$, resulta:

\begin{equation}
%\label{ea.3.10}
\pounds \BRA{\delta(t)} = 1
\end{equation}

Uma importante propriedade da fun\c{c}\~{a}o delta \'{e} a seguite:

\begin{equation}
\label{ea.3.11}
\int_{-\infty}^{\infty}f(t) \delta(t-a)dt=f(a)
\end{equation}

Observa-se que os limites de integra\c{c}\~{a}o na equa\c{c}\~{a}o \ref{ea.3.11} 
podem ser finitos, em cujo caso:

\begin{equation}
%\label{ea.3.12}
\int_{C_{1}}^{C_{2}} f(t) \delta(t-a) dt = \left\{
\begin{array}{ll} 
0 \hspace{5mm} a<C_{1} \mbox{ou} C_{2}<a \\
f(a),C_{1}<a<C_{2}
\end{array}
\right.
\end{equation}

## 4. Translation theorems <a name="section_4"></a> 

Seja:

\begin{equation}
%\label{ea.4.1}
F(t)= f(t) e^{at}
\end{equation}
onde a \'{e} constante complexa. Segundo \ref{ea.2.1}:

\begin{equation}
%\label{ea.4.2}
\overline{F}(s) = \int_{0}^{\infty} e^{-st} f(t) e^{at}dt
\end{equation}

\begin{equation}
%\label{ea.4.3}
\overline{F}(s) = \int_{0}^{\infty} e^{-(s-a)} f(t) dt = \overline{f}(s-a)
\end{equation}

Demonstrou-se, consequentemente, que:

\begin{equation}
%\label{ea.4.4}
\pounds \BRA{f(t) e^{at}} = \overline{f}(s-a)
\end{equation}

Reciprocamente, se na transforma\c{c}\~{a}o:

\begin{equation}
%\label{ea.4.5}
f(s) = \int_{0}^{\infty} e^{-s \lambda} f(\lambda) d \lambda
\end{equation}
introduz-se a troca de vari\'{a}veis:

\begin{equation}
%\label{ea.4.6}
\lambda = t-a
\end{equation}
resulta:

\begin{eqnarray}
%\label{ea.4.7}
\nonumber
\overline{f}(s) &=& \int_{a}^{\infty} e^{-s(t-a)} f(t-a) dt \\
\nonumber
\overline{f}(s) &=& e^{as} \int_{0}^{\infty} e^{-st} \BRA{f(t-a) u(t-a)} dt
\end{eqnarray}               

Deduz-se ent\~{a}o o teorema:

\begin{equation}
%\label{ea.4.8}
\pounds^{-1}\BRA{e^{-as} \overline{f}(s)} = f(t-a) u(t-a)
\end{equation}


## 5. Transformation of derivatives <a name="section_5"></a> 

Seja:

\begin{equation}
\label{ea.5.1}
\pounds \BRA{\frac{df}{dt}} = \int_{0}^{\infty} e^{-st} \frac{df}{dt} dt
\end{equation}

Integrando \ref{ea.5.1} por partes:

\begin{equation}
%\label{ea.5.2}
\pounds \BRA{ \frac{df}{dt}} = e^{-st} f(t) \mid_{0}^{\infty} - \int_{0}^{\infty}
(-s e^{-st})f(t)dt
\end{equation}

Se  $\lim_{t\rightarrow 0} f(t) = f(0)$, ent\~{a}o:

\begin{equation}
%\label{ea.5.3}
\pounds \BRA{ \frac{df}{dt}} = s \overline{f}(s) - f(0)
\end{equation}

Em geral, para a n-\'{e}sima derivada se aplica a equa\c{c}\~{a}o:

\begin{equation}
%\label{ea.5.4}
\pounds \BRA{\frac{ d^{n}f}{d t^{n}}} = s^{n} \overline{f}(s) - s^{n-1}
f(0) - s^{n-2} \frac{df}{dt} \mid_{t=0} = \dots - \frac{ d^{n-1}f}{d t^{n-1}}
\mid_{t=0}
\end{equation}


## 6. Transformation of integrals <a name="section_6"></a> 

Seja a fun\c{c}\~{a}o:

\begin{equation}
%\label{ea.6.1}
f_{1}(t) = \int_{0}^{t} f (\tau) d \tau
\end{equation}

Por defini\c{c}\~{a}o:

\begin{equation}
%\label{ea.6.2}
\pounds \BRA{f_{1}^{(t)}}=\int_{0}^{\infty} e^{-st} (\int_{0}^{t} f(\tau)
d \tau) dt
\end{equation}
logo:

\begin{equation}
%\label{ea.6.3}
\overline{f}(s) = \int_{0}^{\infty} \int_{0}^{t} e^{-st} f(\tau) d\tau dt
\end{equation}

Integrando por partes, sendo:

\begin{equation}
%\label{ea.6.4}
u = - \frac{1}{s} e^{-st} \hspace{20mm} \dot{v} = f(t)
\end{equation}
resulta:

\begin{equation}
%\label{ea.6.5}
\overline{f}_{1} = \BRA{- \frac{1}{s} e^{-st} \int_{0}^{t} f(\tau) 
dt}_{0}^{\infty} + \int_{0}^{\infty} \frac{e^{-st}}{s} f(t) dt
\end{equation}

A express\~{a}o entre colchetes \'{e} nula em ambos limites de 
integra\c{c}\~{a}o, ficando:

\begin{equation}
%\label{ea.6.6}
\pounds \BRA{ \int_{0}^{t} f(\tau) d\tau} = \frac{1}{s} \overline{f} (s)
\end{equation}

Repetindo o procedimento, se demonstra que:

\begin{equation}
%\label{ea.6.7}
\pounds \BRA{ \int_{0}^{t} \int_{0}^{t_{n-1}} \dots \int_{0}^{t_{2}} 
\int_{0}^{t_{1}} f(\tau) d\tau dt_{1} \dots dt_{n-1}} = \frac{1}{s^{n}}
\overline{f}(s)
\end{equation}


## 7. Convolution theorem <a name="section_7"></a> 


Define-se a covolu\c{c}\~{a}o de duas fun\c{c}\~{o}es $f(t)$ e $g(t)$ como 
a integral:

\begin{equation}
%\label{ea.7.1}
f(t) * g(t) \hspace{3mm}\Xi \int_{0}^{t} g(t - \tau) f(\tau) d\tau
\end{equation}
 
Verifica-se facilmente que a covolu\c{c}\~{a}o \'{e} uma opera\c{c}\~{a}o 
comutativa:

\begin{equation}
%\label{ea.7.2}
f(t) * g(t) = g(t) * f(t)
\end{equation}

Equivalentemente:

\begin{equation}
%\label{ea.7.3}
\int_{0}^{t} g(t - \tau) f(\tau) d\tau = \int_{0}^{t} f(t - \tau) g(\tau) 
d\tau
\end{equation}

Al\'{e}m dissso, a covolu\c{c}\~{a}o \'{e} associativa:

\begin{equation}
%\label{ea.7.4}
(f(t) * g(t)) * h(t) = f(t) * (g(t) * h(t))
\end{equation}

Ser\'{a} demonstrado, em covolu\c{c}\~{a}o, de maneira formal, que:

\begin{equation}
%\label{ea.7.5}
\BRA{ f(t) * g(t)} = \overline{f}(s) \overline{g} (s)
\end{equation}

Com efeito:

\begin{equation}
%\label{ea.7.6}
\BRA{f(t) * g(t)} = \int_{0}^{\infty} e^{-st} \int_{0}^{t} f(\tau) 
g(t - \tau) d\tau dt
\end{equation}

O limite superior de integra\c{c}\~{a}o da segunda integral pode ser 
modificado utilizando a propriedade da fun\c{c}\~{a}o salto:

\begin{eqnarray}
%\label{ea.7.7}
\nonumber
g(t - \tau) u(t - \tau) &=& g(t - \tau) \hspace{10mm} \mbox{se}\hspace{2mm} 
\tau<t \\
\nonumber
g(t - \tau) u(t - \tau) &=& 0 \hspace{22mm} \mbox{se} \hspace{2mm} \tau>t \\
\end{eqnarray}
ficando:

\begin{equation}
%\label{ea.7.8}
\BRA{f(t) * g(t)} = \int_{0}^{\infty} e^{-st} \int_{0}^{\infty} f(\tau) 
g(t - \tau) u(t - \tau) d\tau dt
\end{equation}

Trocando agora a ordem de integra\c{c}\~{a}o:

\begin{equation}
%\label{ea.7.9}
\BRA{f(t) * g(t)} = \int_{0}^{\infty} f(\tau) \mid \int_{0}^{\infty} 
g(t - \tau) u(t - \tau) e^{-st} dt \mid d\tau 
\end{equation}
e na express\~{a}o entre colchetes introduz-se a nova vari\'{a}vel 
$t - \tau = v$:

\begin{equation}
%\label{ea.7.10}
\BRA{f(t) * g(t)} = \int_{0}^{\infty} f(\tau) \BRA{\int_{-\tau}^{\infty} 
u(v) g(v) e^{-s(v + \tau)} dv} d\tau
\end{equation}

\begin{equation}
%\label{ea.7.11}
\BRA{f(t) * g(t)} = \int_{0}^{\infty} f(\tau) e^{-st} d\tau \BRA{
\int_{0}^{\infty} g(v) e^{-sv} dv}
\end{equation}

Finalmente:

\begin{equation}
%\label{ea.7.12}
f(t) * g(t) = \overline{f}(s) \overline{g}(s)
\end{equation}

Resultando que resuma o Teorema da Covolu\c{c}\~{a}o da Transforma\c{c}\~{a}o 
de Laplace.


## 8. Solution of differential equations <a name="section_8"></a> 

Let us assume that...
